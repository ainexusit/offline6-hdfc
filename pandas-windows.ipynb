```python
7. Windowing and Rolling Operations
df.rolling() – Apply rolling window operations (e.g., moving average).
df.expanding() – Expanding window operations.
df.ewm() – Exponential weighted moving functions.
df.rolling(window=3).mean() – Calculate rolling mean.
df.rolling(window=3).sum() – Calculate rolling sum.
df.rolling(window=3).std() – Calculate rolling standard deviation.
```

```python
df['sal_roll_3']= df['Salary'].rolling(window=3).sum()
```
